#Python NumPy Tutorial for Beginners

```python
import numpy as np #numpy
```

```python
a=np.array([1,2,3],dtype='int32') #1d array
print(a)
```

```python
b=np.array([[1,2],[2,1]]) #2d array
print(b)
```

```python
a.ndim, b.ndim #dimension
```

```python
a.shape, b.shape #(row,col)
```

```python
a.dtype, b.dtype #datatype
```

```python
a.itemsize, b.itemsize #itemsize (in bytes) = dtype size (in bits) / 8
```

```python
a.size, b.size #count
```

```python
a.nbytes, b.nbytes #nbytes = size * itemsize
```

```python
np.identity(3) #identity matrix
```

```python
#repeat an array
e=np.array([[1,2,3]])
r1=np.repeat(e,3,axis=0)
r1
```

```python
output=np.ones((5,5))
print(output)
z=np.zeros((3,3))
z[1,1]=9
print(z)
output[1:4,1:4]=z
print(output)
```

```python
f=e.copy() #copy
print(f)
f[0,1]=5
print(f)
print(e)
```

```python
#Math
print(f+2)
print(f-2)
print(f*2)
print(f/2)
print(f**2)
print(f+e)
print(np.sin(f))
print(np.cos(f))
```

```python
#linear algebra
g=np.ones((2,3))
h=np.full((3,2),2)
print(g)
print(h)
print(np.matmul(g,h))
```

```python
#determinant
i=np.identity(3)
print(np.linalg.det(i))
```

```python
#statistics
stats=np.array([[1,2,3],[4,5,6]])
stats
```

```python
print(np.min(stats))
print(np.max(stats))
print(np.min(stats,axis=0))
print(np.max(stats,axis=1))
print(np.sum(stats))
print(np.sum(stats,axis=1))
```

```python
#recognizing array
before=np.array([[1,2,3,4],[5,6,7,8]])
print(before)
after=before.reshape((8,1))
print(after)
```

```python
v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])
print(np.vstack((v1,v2,v2)),end="\n\n") #vertically stack
print(np.hstack((v1,v2,v1))) #horizontally stack
```
